```python
from pyspark.sql.functions import *
from pyspark.sql.window import Window
```

```python
df=spark.read.format("delta").load("abfss://bronze@sadevshopify.dfs.core.windows.net/orders")
display(df)
```

```python
df=df.drop('_rescued_data')
display(df)
```

```python
df=df.withColumn("year",year(df.order_date))
df=df.withColumn("month",month(df.order_date))
```

```python
from pyspark.sql.functions import dense_rank
df=df.withColumn("dense_rank", dense_rank().over(Window.partitionBy("year").orderBy(desc("total_amount"))))

display(df)
```

```python
df.write.format("delta").mode("append").saveAsTable("shopify.silver.orders")
```
