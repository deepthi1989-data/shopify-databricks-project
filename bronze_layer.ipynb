```python
dbutils.widgets.text("filename","")
p_filename=dbutils.widgets.get("filename")
```

```python
df=spark.readStream.format("cloudFiles") \
  .option("cloudFiles.format", "parquet") \
  .option("cloudFiles.schemaLocation",f"abfss://bronze@sadevshopify.dfs.core.windows.net/checkpointlocation/schema_{p_filename}")\
  .load(f"abfss://source@sadevshopify.dfs.core.windows.net/{p_filename}")
```

```python
df.writeStream.format("delta")\
.outputMode("append")\
.option("checkpointLocation",f"abfss://bronze@sadevshopify.dfs.core.windows.net/checkpointlocation/{p_filename}")\
.option("path",f"abfss://bronze@sadevshopify.dfs.core.windows.net/{p_filename}")\
.trigger(once=True)\
.start()
```
