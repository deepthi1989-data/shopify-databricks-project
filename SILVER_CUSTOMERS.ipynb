```python
from pyspark.sql.functions import *
```

```python
df=spark.read.format("delta").load("abfss://bronze@sadevshopify.dfs.core.windows.net/customers")
```

```python
df=df.drop("_rescued_data")
```

```python
df=df.withColumn("domain",split(col("email"),"@")[1])
```

```python
df.groupBy("domain").agg(count("customer_id").alias("domain_count")).orderBy(desc("domain_count"))
```

```python
df=df.withColumn("full_name",concat(col("first_name"),lit(" "),col("last_name")))
df=df.drop("first_name","last_name")
```

```python
df.write.mode("overwrite").format("delta").saveAsTable("shopify.silver.customers")
```
