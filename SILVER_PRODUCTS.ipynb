```python
df=spark.read.format("delta").load("abfss://bronze@sadevshopify.dfs.core.windows.net/products")
display(df)
```

```python
df.drop('_rescued_data')
```

```python
df.createOrReplaceTempView("products")
```

```python
%sql
create or replace function shopify.bronze.discounted_price(p_price double)
  returns double
  return p_price * 0.9
```

```python
%sql
select product_id, shopify.bronze.discounted_price(price) as discounted_price
from products
```

```python
df.write.format("delta").mode("overwrite").saveAsTable("shopify.silver.products")
```
